```python
def getAreas(name):
    usedf = propertyCode[propertyCode['Building Use'].str.contains(name, na=False)]
    uselst = []
    unames = electric['Meter'].unique()
    a = {'Meter': unames}
    unames = pd.DataFrame(a)
    ucodelist = [ab[0] for ab in unames['Meter'].str.split()]
    ucodes = pd.Series(ucodelist)
    for i in usedf['Property Code']:
        uselst.append(i)
    groupdf = electric[codes.isin(uselst)]
    bldlst = pd.Series([b[0] for b in electric['Meter'].str.split()])
    bldlst = bldlst.unique()
    sbldlst1 = []
    sbldlst = []
    nalst = []
    narea = pd.DataFrame()
    for i in bldlst:
        sgroupdf = groupdf[groupdf['Meter'].str.contains(i)]
        if not sgroupdf.empty:
            sbldlst1.append(list(b[0] for b in sgroupdf['Meter'].str.split()))
            sbldlst = [item[0] for item in sbldlst1]
        for i in sbldlst:
            nalst.append(propertyCode[propertyCode['Property Code'].isin([i])]['FCIM Gross Area'])
            narea = pd.DataFrame(nalst).sum(axis=1)
            narea = narea.reset_index(drop=True)
            narea = narea.unique()
        data = {'Meter': sbldlst, 'Gross Area': narea}
        areadf= pd.DataFrame(data)
        for i in areadf['Meter']:
            nname = str(unames[ucodes.isin([i])]['Meter'][-1:])
            nname = nname[3:-27]
            areadf = areadf.replace(i, nname)
    return areadf
```

```python
areadf = getAreas("Dorm")
```

```python
def normUsage(df, area):
    df['Normalized Usage'] = df['Usage'].divide(area)
```

```python
ddf1 = normUsage(ddf1, 50697.42 )
ddf2 = normUsage(ddf2, 121452.48 )
ddf3 = normUsage(ddf3, 338679.33 )
ddf4 = normUsage(ddf4, 201402.16)
ddf5 = normUsage(ddf5, 237120.34)
ddf6 = normUsage(ddf6, 95682.00 )
```

```python
ax1 = ddf1.plot(x='Temperature', y='Normalized Usage', kind='scatter', color='red', s=0.5, label='Breckinridge', figsize=(20, 10))
ddf2.plot(ax=ax1, x='Temperature', y='Normalized Usage', kind='scatter', color='orange', s=0.5, label="NGRH")
ddf3.plot(ax=ax1, x='Temperature', y='Normalized Usage', kind='scatter', color='yellow', s=0.5, label="6031 S. Ellis")
ddf4.plot(ax=ax1, x='Temperature', y='Normalized Usage', kind='scatter', color='green', s=0.5, label="Burton Judson")
ddf5.plot(ax=ax1, x='Temperature', y='Normalized Usage', kind='scatter', color='blue', s=0.5, label=" International House")
ddf6.plot(ax=ax1, x='Temperature', y='Normalized Usage', kind='scatter', color='black', s=0.5, label="Max Palevsky Halls")
plt.title('Temperature vs Usage (Dorms)')
        
```
