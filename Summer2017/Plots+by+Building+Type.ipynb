```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('Full Eletric Interval 042016.csv', skiprows = 4)
```

```python
df
```

```python
df2 = pd.read_excel('SIMS_Property_List_4-6-2016.xlsx')
```

```python
df2
```

```python
codelst = [b[0] for b in df['Meter'].str.split()]
```

```python
codes = pd.Series(codelst)
```

```python
dorms = []
for i in range(879):
    if df2['Building Use'][i]=='Dormitory':
        dorms.append(df2['Property Code'][i])
```

```python
dorms = pd.Series(dorms)
```

```python
dormdf = df[codes.isin(dorms)]
```

```python
dormdf.plot(x='Meter', y='Usage', figsize=(40,10))
```
