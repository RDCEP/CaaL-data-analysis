## Get all codes, then unique codes


```python
codes = pd.Series([b[0] for b in e2['Meter'].str.split()])
```

```python
unicodes = codes.unique()
unicodes = pd.Series(unicodes)
```

## Get cumulative usage

```python
clst = []
for i in unicodes:
    #for each unique code, find its Usage value in master df and add them all together, then append to list
    clst.append(e2.Usage[codes.isin([i])].sum())
```

## New dataframe w/ just cumulative

```python
data = {'Meter': unicodes, 'Cumulative Usage': clst}
cdf = pd.DataFrame(data)
```

```python
namelst = []
for i in unicodes:
    namelst.append(pc2[pc2['Property Code'].isin([i])]['Property Name'])
```

```python
namelst
name = pd.DataFrame(namelst)
name = name.replace(np.nan, '', regex=True)
name = name.reset_index(drop=True)
name =name.sum(axis=1)
name = pd.DataFrame(name)
name.columns = ['Full Name']
```

```python
cdf['Full Name'] = cdf['Meter'] + " " + name['Full Name']
cdf
```

```python
ccodes =[]
for i in cdf['Meter']:
    ccodes.append(i)
ccodes = pd.Series(ccodes)
```

```python
def getAvg(name):
    cumulative = name['Cumulative Usage'].sum()
    avg = cumulative/(len(name['Cumulative Usage']))
    return avg
```

## Top 15 (normalized for area)

```python
cadf = cdf
cacodes = pd.Series([b[0] for b in cdf['Full Name'].str.split()])
cacodes = cacodes.unique()
```

#### Get area data

```python
arealst = []
for b in cacodes:
    arealst.append(pc2[pc2['Property Code'].isin([b])]['FCIM Gross Area'])
area = pd.DataFrame(arealst).sum(axis=1)
area = area.reset_index(drop=True)
```

```python
cadf['Cumulative Usage (area)'] = cadf['Cumulative Usage'].divide(area)
cadf = cadf.replace(np.inf, np.nan)
```

```python
cadf = cadf.fillna(0)
```

#### Get percentages, normalized for area

```python
aproplst = []
atot = cadf['Cumulative Usage (area)'].sum()

for i in cadf['Cumulative Usage (area)']:
    aproplst.append(i/atot)
```

```python
cadf['Proportions (area)'] = aproplst
```

```python
cadf2 = cadf.sort_values(by ='Proportions (area)', ascending=False)
```

#### Get top 15 buildings

```python
cadf2 = cadf2.reset_index(drop=True)
cadf3 = cadf2.loc[0:14]
cadf3
```

#### Get ages of buildings

```python
agelst = []

for i in cadf3['Meter']:
    agelst.append(pc2[pc2['Property Code'].isin([i])]['Year Constructed'])
```

```python
age = pd.DataFrame(agelst).sum(axis=1)
age = age.reset_index(drop=True)
```

```python
cadf3['Year'] = age
```

```python
cadf3
```

```python
cadf3.plot(x='Year', y='Cumulative Usage (area)', kind='scatter', figsize=(10,5))
plt.title('Cumulative usage vs age (area)')
plt.savefig('Graphs/Cumulative usage (area) v age.png')
```
