```python
def getAreas(name):
    usedf = propertyCode[propertyCode['Building Use'].str.contains(name, na=False)]
    uselst = []
    unames = electric['Meter'].unique()
    a = {'Meter': unames}
    unames = pd.DataFrame(a)
    ucodelist = [ab[0] for ab in unames['Meter'].str.split()]
    ucodes = pd.Series(ucodelist)
    for i in usedf['Property Code']:
        uselst.append(i)
    groupdf = electric[codes.isin(uselst)]
    bldlst = pd.Series([b[0] for b in electric['Meter'].str.split()])
    bldlst = bldlst.unique()
    sbldlst1 = []
    sbldlst = []
    nalst = []
    narea = pd.DataFrame()
    for i in bldlst:
        sgroupdf = groupdf[groupdf['Meter'].str.contains(i)]
        if not sgroupdf.empty:
            sbldlst1.append(list(b[0] for b in sgroupdf['Meter'].str.split()))
            sbldlst = [item[0] for item in sbldlst1]
        for i in sbldlst:
            nalst.append(propertyCode[propertyCode['Property Code'].isin([i])]['FCIM Gross Area'])
            narea = pd.DataFrame(nalst).sum(axis=1)
            narea = narea.reset_index(drop=True)
            narea = narea.unique()
        data = {'Meter': sbldlst, 'Gross Area': narea}
        areadf= pd.DataFrame(data)
        for i in areadf['Meter']:
            nname = str(unames[ucodes.isin([i])]['Meter'][-1:])
            nname = nname[3:-27]
            areadf = areadf.replace(i, nname)
    return areadf
```
