```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('Full Eletric Interval 042016.csv', skiprows = 4)
```

```python
df
```

```python
df.plot(x='Meter', y='Usage', figsize=(40,10))
```
